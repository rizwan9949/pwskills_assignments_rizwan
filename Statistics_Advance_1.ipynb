## Theoretical Questions

### 1. What is a random variable in probability theory?
A random variable is a variable whose possible values are numerical outcomes of a random phenomenon. It can be discrete (finite or countable outcomes) or continuous (infinite outcomes within a range).

### 2. What are the types of random variables?
- **Discrete Random Variables**: Take on a finite or countably infinite set of values.
- **Continuous Random Variables**: Take any value within a given range.

### 3. What is the difference between discrete and continuous distributions?
Discrete distributions have specific, countable outcomes (e.g., rolling a die), while continuous distributions have outcomes over a range (e.g., height measurements).

### 4. What are Probability Distribution Functions (PDF)?
PDFs describe the probabilities of possible outcomes of a random variable. For discrete variables, it's called a Probability Mass Function (PMF), and for continuous variables, it's a Probability Density Function.

### 5. How do Cumulative Distribution Functions (CDF) differ from PDFs?
CDF gives the cumulative probability up to a certain value, while PDF provides the probability density at a specific point.

### 6. What is a discrete uniform distribution?
A discrete uniform distribution assigns equal probabilities to a finite set of outcomes (e.g., rolling a fair die).

### 7. Key properties of a Bernoulli distribution:
- Only two outcomes: success (1) or failure (0).
- Parameter $ p $: probability of success.
- Mean = $ p $, Variance = $ p(1-p) $.

### 8. What is the binomial distribution, and how is it used?
It models the number of successes in $ n $ independent Bernoulli trials with success probability $ p $. Used in scenarios like coin flips.

### 9. What is the Poisson distribution and where is it applied?
Models the number of events occurring in a fixed interval of time or space. Applied in call centers, traffic flow analysis, etc.

### 10. What is a continuous uniform distribution?
All values in a given range are equally likely. Example: Randomly picking a number between 0 and 1.

### 11. Characteristics of a normal distribution:
Symmetric, bell-shaped curve with mean = median = mode. Defined by mean ($ \mu $) and standard deviation ($ \sigma $).

### 12. What is the standard normal distribution?
A normal distribution with mean = 0 and standard deviation = 1. Critical for standardizing data.

### 13. What is the Central Limit Theorem (CLT)?
The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the original distribution.

### 14. How does CLT relate to the normal distribution?
CLT ensures that even non-normal data can be approximated by a normal distribution for large sample sizes.

### 15. Application of Z statistics in hypothesis testing:
Z-scores standardize data to compare it to a standard normal distribution, enabling hypothesis testing.

### 16. How do you calculate a Z-score?
$ Z = \frac{X - \mu}{\sigma} $, where $ X $ is the data point, $ \mu $ is the mean, and $ \sigma $ is the standard deviation.

### 17. What are point estimates and interval estimates?
- Point estimate: Single value estimate of a population parameter.
- Interval estimate: Range of values (confidence interval) around the point estimate.

### 18. Significance of confidence intervals:
They provide a range of plausible values for a population parameter with a specified confidence level.

### 19. Relationship between Z-score and confidence interval:
Z-scores determine the critical values for constructing confidence intervals.

### 20. How are Z-scores used to compare different distributions?
By standardizing data, Z-scores allow comparison across distributions with different means and standard deviations.


## 21. Assumptions for Central Limit Theorem (CLT):
- Large sample size ($ n \geq 30 $).
- Independent and identically distributed (i.i.d.) samples.
- Finite mean and variance in the population.

## 22. Expected Value:
The expected value is the weighted average of all outcomes, weighted by their probabilities:
$$
E(X) = \sum x_i \cdot P(X = x_i)
$$

## 23. Probability Distribution and Expected Outcome:
A probability distribution shows possible outcomes and their probabilities. The expected outcome is the long-term average derived from this distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson, norm, uniform
```

```python
# 1. Generate a random variable and display its value
random_var = np.random.randint(1, 10)
print(f"Random Variable Value: {random_var}")
```

```python
# 2. Generate a discrete uniform distribution and plot PMF
values = np.arange(1, 7)  # Rolling a die
pmf = np.ones(len(values)) / len(values)
plt.bar(values, pmf)
plt.title("Discrete Uniform Distribution (PMF)")
plt.show()
```

```python
# 3. Calculate PDF of a Bernoulli distribution
def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

p = 0.7
x_values = [0, 1]
pdf_values = [bernoulli_pdf(p, x) for x in x_values]
plt.bar(x_values, pdf_values)
plt.title("Bernoulli Distribution (PDF)")
plt.show()
```

```python
# 4. Simulate a binomial distribution and plot histogram
n, p = 10, 0.5
binomial_data = np.random.binomial(n, p, 1000)
plt.hist(binomial_data, bins=range(0, n+2), align='left', rwidth=0.8)
plt.title("Binomial Distribution (Histogram)")
plt.show()
```

```python
# 5. Create a Poisson distribution and visualize it
lambda_ = 3
poisson_data = np.random.poisson(lambda_, 1000)
plt.hist(poisson_data, bins=range(0, 11), align='left', rwidth=0.8)
plt.title("Poisson Distribution")
plt.show()
```

```python
# 6. Plot CDF of a discrete uniform distribution
cdf = np.cumsum(pmf)
plt.step(values, cdf, where='post')
plt.title("Discrete Uniform Distribution (CDF)")
plt.show()
```

```python
# 7. Generate a continuous uniform distribution and visualize it
uniform_data = np.random.uniform(0, 1, 1000)
plt.hist(uniform_data, bins=20, edgecolor='black')
plt.title("Continuous Uniform Distribution")
plt.show()
```

```python
# 8. Simulate data from a normal distribution and plot histogram
normal_data = np.random.normal(0, 1, 1000)
plt.hist(normal_data, bins=20, edgecolor='black')
plt.title("Normal Distribution (Histogram)")
plt.show()
```

```python
# 9. Calculate and plot Z-scores
data = np.random.normal(50, 10, 100)
z_scores = (data - np.mean(data)) / np.std(data)
plt.hist(z_scores, bins=20, edgecolor='black')
plt.title("Z-Scores")
plt.show()
```

```python
# 10. Implement CLT for a non-normal distribution
sample_means = [np.mean(np.random.exponential(1, 100)) for _ in range(1000)]
plt.hist(sample_means, bins=20, edgecolor='black')
plt.title("Central Limit Theorem (CLT)")
plt.show()
```

```python
# 11. Write a Python program to compare the standard deviations of two datasets and visualize the difference
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(50, 15, 100)

std1, std2 = np.std(data1, ddof=1), np.std(data2, ddof=1)
print(f"Std Dev Dataset 1: {std1:.2f}, Std Dev Dataset 2: {std2:.2f}")

plt.boxplot([data1, data2], labels=["Dataset 1", "Dataset 2"])
plt.title("Comparison of Standard Deviations")
plt.show()
```

```python
# 12. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
data = np.random.normal(50, 10, 100)
range_ = np.max(data) - np.min(data)
q1, q3 = np.percentile(data, 25), np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {range_:.2f}, IQR: {iqr:.2f}")
```

```python
# 13. Implement Z-score normalization on a dataset and visualize its transformation
data = np.random.normal(50, 10, 100)
z_scores = (data - np.mean(data)) / np.std(data)

plt.hist(data, bins=20, alpha=0.5, label="Original Data")
plt.hist(z_scores, bins=20, alpha=0.5, label="Z-Score Normalized Data")
plt.title("Z-Score Normalization")
plt.legend()
plt.show()
```

```python
# 14. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution
from scipy.stats import skew, kurtosis

data = np.random.normal(50, 10, 100)
skewness, kurt = skew(data), kurtosis(data)
print(f"Skewness: {skewness:.2f}, Kurtosis: {kurt:.2f}")
```

```python
# 15. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
sample_size, num_samples = 50, 1000
data = np.random.exponential(1, (num_samples, sample_size))
sample_means = np.mean(data, axis=1)

plt.hist(sample_means, bins=30, edgecolor='black', color='skyblue')
plt.title("CLT Verification")
plt.show()
```

```python
# 16. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, loc=0, scale=1)
plt.plot(x, pdf, label="Standard Normal", color='blue')
plt.title("Standard Normal Distribution")
plt.legend()
plt.show()
```

```python
# 17. Generate random variables and calculate their corresponding probabilities using the binomial distribution
n, p = 10, 0.5
x_values = np.arange(0, n+1)
probabilities = binom.pmf(x_values, n, p)

plt.bar(x_values, probabilities, color='orange', edgecolor='black')
plt.title("Binomial PMF")
plt.show()
```

```python
# 18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
data_point, mean, std_dev = 70, 50, 10
z_score = (data_point - mean) / std_dev

x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, loc=0, scale=1)
plt.plot(x, pdf, color='blue')
plt.axvline(z_score, color='red', linestyle='--')
plt.title("Z-Score on Standard Normal")
plt.show()
```

```python
# 19. Implement hypothesis testing using Z-statistics for a sample dataset
sample_mean, pop_mean, pop_std, n = 52, 50, 10, 100
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
critical_value = norm.ppf(0.975)

if abs(z_stat) > critical_value:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

```python
# 20. Create a confidence interval for a dataset
data = np.random.normal(50, 10, 100)
mean, std_dev = np.mean(data), np.std(data, ddof=1)
margin = norm.ppf(0.975) * (std_dev / np.sqrt(len(data)))
lower, upper = mean - margin, mean + margin
print(f"Confidence Interval: ({lower:.2f}, {upper:.2f})")
```
