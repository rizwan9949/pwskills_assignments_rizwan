# Theoretical Questions

# 1. What is K-Nearest Neighbors (KNN) and how does it work?
K-Nearest Neighbors (KNN) is a supervised machine learning algorithm used for classification and regression tasks.
It works by finding the K closest data points (neighbors) to a given query point in the feature space and making predictions based on their labels or values.
For classification, the majority class among the neighbors is assigned to the query point.
For regression, the average or weighted average of the neighbors' values is used as the prediction.


# 2. What is the difference between KNN Classification and KNN Regression?

In KNN Classification, the output is a class label determined by the majority vote of the K nearest neighbors.
In KNN Regression, the output is a continuous value calculated as the average (or weighted average) of the K nearest neighbors' target values.


# 3. What is the role of the distance metric in KNN?

The distance metric determines how the similarity between data points is measured. Common metrics include:
- Euclidean Distance: Measures straight-line distance between points.
- Manhattan Distance: Measures sum of absolute differences along each dimension.
- Minkowski Distance: Generalized form of Euclidean and Manhattan distances.
Choosing the right metric depends on the data and problem context.


# 4. What is the Curse of Dimensionality in KNN?

The Curse of Dimensionality refers to the challenges that arise when working with high-dimensional data. In KNN, as the number of dimensions increases:
- Data points become sparse, making distances less meaningful.
- Computational complexity increases due to the need to compute distances across many dimensions.
This can degrade KNN's performance unless dimensionality reduction techniques (e.g., PCA) are applied.


# 5. How can we choose the best value of K in KNN?

The optimal value of K depends on the dataset and can be chosen using:
- Cross-validation: Evaluate KNN performance for different K values and select the one with the best accuracy or lowest error.
- Rule of Thumb: Start with K = sqrt(N), where N is the number of samples.
- Domain Knowledge: Use insights about the problem to guide the choice of K.
Smaller K values make the model sensitive to noise, while larger K values smooth out decision boundaries.


# 6. What are KD Tree and Ball Tree in KNN?

KD Tree and Ball Tree are data structures used to speed up neighbor searches in KNN:
- KD Tree: Recursively partitions the data into axis-aligned regions, allowing efficient search in low-to-moderate dimensions.
- Ball Tree: Partitions the data into hyperspheres, which is more efficient in higher dimensions.
Both reduce the computational cost compared to brute-force search.


# 7. When should you use KD Tree vs. Ball Tree?

Use KD Tree for low-dimensional data (<20 dimensions) where axis-aligned splits are effective.
Use Ball Tree for high-dimensional data where hyperspherical partitions perform better.
Brute-force search may still be preferable for very small datasets or extremely high dimensions.


# 8. What are the disadvantages of KNN?

Disadvantages of KNN include:
- High computational cost for large datasets due to distance calculations.
- Sensitivity to irrelevant or noisy features.
- Requires careful choice of K and distance metric.
- Performance degrades with high-dimensional data (Curse of Dimensionality).
- Does not handle missing values natively.


# 9. How does feature scaling affect KNN?

Feature scaling ensures that all features contribute equally to distance calculations. Without scaling, features with larger ranges dominate the distance metric, leading to biased results. Common scaling methods include Min-Max Scaling and Standardization.


# 10. What is PCA (Principal Component Analysis)?

PCA is a dimensionality reduction technique that transforms data into a new coordinate system where the axes (principal components) capture the maximum variance in the data. It reduces redundancy and noise while preserving important patterns.


# 11. How does PCA work?

PCA works by:
1. Standardizing the data to have zero mean and unit variance.
2. Computing the covariance matrix of the data.
3. Finding the eigenvalues and eigenvectors of the covariance matrix.
4. Sorting the eigenvectors by their corresponding eigenvalues (variance explained).
5. Projecting the data onto the top K eigenvectors to reduce dimensions.


# 12. What is the geometric intuition behind PCA?

Geometrically, PCA rotates the data to align it with the directions of maximum variance. The first principal component captures the direction with the largest spread, the second captures the next largest orthogonal direction, and so on.

# 13. What are Eigenvalues and Eigenvectors in PCA?
Eigenvalues represent the amount of variance captured by each principal component. Eigenvectors define the directions (axes) of these components in the original feature space.


# 14. What is the difference between Feature Selection and Feature Extraction?

- Feature Selection: Selects a subset of the original features based on importance.
- Feature Extraction: Transforms the original features into a new set of features (e.g., PCA).
Feature extraction creates new representations, while feature selection retains the original features.


# 15. How do you decide the number of components to keep in PCA?

The number of components can be decided based on:
- Explained Variance Ratio: Retain enough components to explain a desired percentage (e.g., 95%) of the total variance.
- Scree Plot: Look for an "elbow point" where additional components contribute little variance.


# 16. Can PCA be used for classification?

PCA itself is not a classification algorithm but can be used as a preprocessing step to reduce dimensionality and improve classification performance by removing noise and redundancy.


# 17. What are the limitations of PCA?

Limitations of PCA include:
- Assumes linear relationships between variables.
- Sensitive to feature scaling.
- Ignores class labels, potentially losing discriminative information.
- May not perform well with non-linear data.


# 18. How do KNN and PCA complement each other?

PCA reduces the dimensionality of the data, addressing the Curse of Dimensionality in KNN. This improves computational efficiency and can enhance KNN's performance by focusing on the most informative features.

# 19. How does KNN handle missing values in a dataset?
KNN does not natively handle missing values. Missing values must be imputed before training, either using simple methods (mean/median imputation) or advanced techniques like KNN Imputation.

# 20. What are the key differences between PCA and Linear Discriminant Analysis (LDA)?
- PCA: Unsupervised; focuses on maximizing variance without considering class labels.
- LDA: Supervised; focuses on maximizing class separability by maximizing between-class variance and minimizing within-class variance.

```python
# 21. Train a KNN Classifier on the Iris dataset and print model accuracy
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print("KNN Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# 22. Train a KNN Regressor on a synthetic dataset and evaluate using Mean Squared Error (MSE)
from sklearn.datasets import make_regression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

X, y = make_regression(n_samples=100, n_features=1, noise=0.1, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

knn_reg = KNeighborsRegressor(n_neighbors=5)
knn_reg.fit(X_train, y_train)
y_pred_reg = knn_reg.predict(X_test)
print("KNN MSE:", mean_squared_error(y_test, y_pred_reg))
```

```python
# 23. Train a KNN Classifier using different distance metrics (Euclidean and Manhattan) and compare accuracy
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)
knn_euclidean = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn_manhattan = KNeighborsClassifier(n_neighbors=3, metric='manhattan')
knn_euclidean.fit(X_train, y_train)
knn_manhattan.fit(X_train, y_train)

y_pred_euclidean = knn_euclidean.predict(X_test)
y_pred_manhattan = knn_manhattan.predict(X_test)

print("Euclidean Accuracy:", accuracy_score(y_test, y_pred_euclidean))
print("Manhattan Accuracy:", accuracy_score(y_test, y_pred_manhattan))
```

```python
# 24. Train a KNN Classifier with different values of K and visualize decision boundaries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=100, n_features=2, n_classes=2, n_informative=2, n_redundant=0, random_state=42)
K_values = [1, 5, 10]

for i, K in enumerate(K_values):
    knn_boundary = KNeighborsClassifier(n_neighbors=K)
    knn_boundary.fit(X, y)
    xx, yy = np.meshgrid(np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 100),
                         np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 100))
    Z = knn_boundary.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.subplot(1, len(K_values), i+1)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
    plt.title(f"K={K}")
plt.show()
```

```python
# 25. Apply Feature Scaling before training a KNN model and compare results with unscaled data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(iris.data)
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, iris.target, test_size=0.3, random_state=42)

knn_scaled = KNeighborsClassifier(n_neighbors=3)
knn_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = knn_scaled.predict(X_test_scaled)
print("Scaled Data Accuracy:", accuracy_score(y_test, y_pred_scaled))
```
